```python
from wikipedia_scraper import fetch_wikipedia_page, search_wikipedia
```

```python
name_list = search_wikipedia("1990 Rugby Women's RugbyFest 1990")
```

```python
for name in name_list:
    if '1990' in name and 'rugby' in name.lower() and 'RugbyFest 1990' in name:
        if name[-len('RugbyFest 1990'):] == 'RugbyFest 1990':
            url = name
```

```python
page = fetch_wikipedia_page(url)
```
